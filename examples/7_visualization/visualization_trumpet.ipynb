# Power and Trumpet plot

```python
import gwaslab as gl
```

## Calculate power using .get_power

```python
gl.get_power(mode="b", 
             ncase=1000,
             ncontrol=11000, 
             beta=0.3, 
             prevalence=0.1, 
             daf=0.2)
```

```python
gl.get_power(mode="q", eaf= 0.2, beta=0.3, n=1000)
```

## Plot power (mode="b")

```python
a = gl.plot_power_x(mode="b",
                    x="MAF",
                    ncases=1000, 
                    ncontrols=[11000,15000,110000],
                    betas=0.3,
                    prevalences=0.1,
                    xscale="nonlog")
```

```python
a = gl.plot_power_x(mode="b",
                    mafs=[0.5,0.4,0.3,0.2,0.9,0.1], 
                    betas= 0.1, 
                    ncontrols=10000, 
                    prevalences=0.1, 
                    xscale="nonlog")
```

```python
a = gl.plot_power_x(mode="b",
                    betas=[0.5,0.4,0.3,0.2], 
                    mafs= 0.1,
                    xscale="nonlog")
```

## Trumpet plot for quantitative traits

```python
#Sample data : Akiyama, M., Ishigaki, K., Sakaue, S., Momozawa, Y., Horikoshi, M., Hirata, M., ... & Kamatani, Y. (2019). Characterizing rare and low-frequency height-associated variants in the Japanese population. Nature communications, 10(1), 4393.
```

```python
mysumstats = gl.Sumstats("bbj_height.tsv",fmt="auto",eaf="MAF",sep="\s+",other=["Gene"],n=160000,verbose=False)
mysumstats.basic_check(verbose=False)
```

```python
mysumstats.data
```

```python
a = mysumstats.plot_trumpet(mode="q",
                            ts=[0.2,0.4,0.6,0.8] ,
                            anno="Gene",
                            anno_style="expand",
                            cmap="cool",
                            sig_level=5e-8,
                            build="19",
                            anno_x=0.01,
                            anno_y=1,
                            p_level=5e-8,
                            n_matrix=2000,
                            fontsize=12,
                            xscale="log",
                            repel_force=0.15,
                            yscale_factor=5.63,
                            sort="eaf",
                            ylim=(-6,4),
                            save=True)
```

```python
a = mysumstats.plot_trumpet(mode="q",
                            ts=[0.8] ,
                            anno="Gene",
                            anno_style="expand",
                            sig_level=5e-8,
                            build="19",
                            anno_x=0.01,
                            anno_y=1,
                            p_level=5e-8,
                            n_matrix=2000,
                            fontsize=12,
                            cmap="cool",
                            xscale="nonlog",
                            repel_force=0.05,
                            yscale_factor=5.63,
                            sort="beta",
                            ylim=(-6,4),
                            markercolor="#349beb")
```

## Trumpet plot for binary traits

```python
mysumstats =gl.Sumstats("t2d_bbj.txt.gz",fmt="auto",verbose=False)
mysumstats.basic_check(verbose=False)
```

```python
a = mysumstats.plot_trumpet(mode="b",
                            ncase=36614,
                            ncontrol=155150,
                            ts=[0.2,0.4,0.6,0.8] ,
                            anno="GENENAME",
                            anno_style="expand",
                            cmap="cool",
                            sig_level=5e-8,
                            build="19",
                            anno_x=0.01,
                            anno_y=0.2,
                            p_level=5e-8,
                            n_matrix=2000,
                            fontsize=12,
                            xscale="log",
                            repel_force=0.15,
                            sort="eaf",
                            ylim=(-3,3),
                            save=True)
```
