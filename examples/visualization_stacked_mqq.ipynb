# Load gwaslab

```python
import sys
sys.path.insert(0,"/home/yunye/work/gwaslab/src")
import gwaslab as gl
```

# Load sumstats

```python
gl1 = gl.Sumstats("bbj_bmi_female.txt.gz",snpid="SNP",chrom="CHR",ea="ALT",nea="REF",pos="POS",p="P",sep="\t")
gl2 = gl.Sumstats("bbj_bmi_male.txt.gz",snpid="SNP",chrom="CHR",ea="ALT",nea="REF",pos="POS",p="P",sep="\t")
```

# basic check

```python
gl1.basic_check()
gl2.basic_check()
```

# stacked manhattan plot

```python
gl.plot_stacked_mqq(objects=[gl1,gl2],
                    mode="m",
                    skip=3,
                    anno=True,
                    anno_style="tight",
                    sig_level_lead=1e-20,
                    titles=["Male","Female"],
                    anno_args={"rotation":0})
```

# stacked regional plot

```python
gl.plot_stacked_mqq(objects=[gl1,gl2],
                    vcfs=[gl.get_path("1kg_eas_hg19")],
                    region=(19,46214297 - 500000,46214297 + 500000),
                    mode="r",
                    build="19",
                    anno=True,
                    anno_style="tight",
                    titles=["Male","Female"],
                    title_args={"size":20},
                    anno_args={"rotation":0})
```
