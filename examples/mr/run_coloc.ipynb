```python
mysumstats = gl.Sumstats("./koges_bmi.tsv.gz", 
                             chrom="chrom",
                             pos="pos",
                             ea="alt",
                             nea="ref",
                             eaf="af",
                             beta="beta",
                             se="sebeta",
                             p="pval",nrows=1000000,
                            verbose=False) 
```

```python
mysumstats.basic_check(verbose=True)
```

```python
dirty_sumstats = gl.Sumstats("../t2d_bbj.txt.gz", 
                             snpid="SNP",
                             chrom="CHR",
                             pos="POS",
                             ea="REF",
                             nea="ALT",
                             eaf="Frq",
                             beta="BETA",
                             se="SE",
                             p="P",
                             nrows=1000000,
                             verbose=False) 
```

```python
dirty_sumstats.basic_check(verbose=False)
```

```python
pair = gl.SumstatsPair(mysumstats, dirty_sumstats)
```
