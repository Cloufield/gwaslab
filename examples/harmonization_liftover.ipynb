# Liftover

```python
import gwaslab as gl
```

```python
mysumstats = gl.Sumstats("t2d_bbj.txt.gz",
             snpid="SNP",
             chrom="CHR",
             pos="POS",
             ea="ALT",
             nea="REF",
             neaf="Frq",
             beta="BETA",
             se="SE",
             p="P",nrows=500000)
```

```python
mysumstats.basic_check()
```

```python
mysumstats.data
```

```python
mysumstats.liftover(n_cores=3, from_build="19", to_build="38")
```
