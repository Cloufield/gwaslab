# trumpet q

```python
#Sample data : Akiyama, M., Ishigaki, K., Sakaue, S., Momozawa, Y., Horikoshi, M., Hirata, M., ... & Kamatani, Y. (2019). Characterizing rare and low-frequency height-associated variants in the Japanese population. Nature communications, 10(1), 4393.
```

```python
mysumstats = gl.Sumstats("bbj_height.tsv",fmt="auto",eaf="MAF",sep="\s+",other=["Gene"],n=160000)
```

```python
mysumstats.basic_check()
```

```python
mysumstats.data
```

```python
a = mysumstats.plot_trumpet(mode="q",
                            ts=[0.2,0.4,0.6,0.8] ,
                            anno="Gene",
                            anno_style="expand",
                            cmap="cool",
                            sig_level=5e-8,
                            build="19",
                            anno_x=0.01,
                            anno_y=1,
                            p_level=5e-8,
                            n_matrix=2000,
                            fontsize=12,
                            xscale="log",
                            repel_force=0.15,
                            yscale_factor=5.63,
                            sort="eaf",
                            ylim=(-5.5,4),
                            save=True)
```

```python
a = mysumstats.plot_trumpet(mode="q",
                            ts=[0.8] ,
                            anno="Gene",
                            anno_style="expand",
                            sig_level=5e-8,
                            build="19",
                            anno_x=0.01,
                            anno_y=1,
                            p_level=5e-8,
                            n_matrix=2000,
                            fontsize=12,
                            cmap="cool",
                            xscale="nonlog",
                            repel_force=0.05,
                            yscale_factor=5.63,
                            sort="beta",
                            ylim=(-5.5,4),
                            markercolor="#349beb")
```
