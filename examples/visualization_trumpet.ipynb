```python
#Sample data : Akiyama, M., Ishigaki, K., Sakaue, S., Momozawa, Y., Horikoshi, M., Hirata, M., ... & Kamatani, Y. (2019). Characterizing rare and low-frequency height-associated variants in the Japanese population. Nature communications, 10(1), 4393.
```

```python
mysumstats =gl.Sumstats("height_power.txt",
             snpid="rsID",
             chrom="CHR",
             pos="POS",
             eaf="MAF",
             beta="BETA",
             p="P",
             n=170000,
             se="SE",
             other=["Gene"],
             nrows=500000)
```

```python
mysumstats.basic_check()
```

```python
a = mysumstats.plot_trumpet(mode="q",
                            ts=[0.2,0.4,0.6,0.8] ,
                            anno="Gene",
                            anno_style="expand",
                            cmap="cool",
                            sig_level=5e-8,
                            build="19",
                            anno_x=0.01,
                            anno_y=1,
                            p_level=5e-8,
                            n_matrix=2000,
                            fontsize=12,
                            xscale="log",
                            repel_force=0.15,
                            yscale_factor=5.63,
                            sort="eaf",
                            ylim=(-5,4),
                            save=True)
```
