```python
mysumstats = gl.Sumstats("t2d_bbj.txt.gz",
             snpid="SNP",
             chrom="CHR",
             pos="POS",
             ea="ALT",
             nea="REF",
             neaf="Frq",
             beta="BETA",
             se="SE",
             p="P",
             direction="Dir",
             n="N")
```

```python
mysumstats.plot_mqq(skip=3,cut=20, mode="mqq",stratified=True)
```

```python
mysumstats.filter_in(eq={"CHR":"7"})
mysumstats.basic_check()
```
