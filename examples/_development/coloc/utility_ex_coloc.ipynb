```python
mysumstats = gl.Sumstats("../../0_sample_data/koges_bmi.tsv.gz", 
                             chrom="chrom",
                             pos="pos",
                             ea="alt",
                             nea="ref",
                             eaf="af",
                             beta="beta",
                             se="sebeta",
                             p="pval",nrows=1000000,
                            verbose=False) 
```

```python
mysumstats.data
```

```python
mysumstats.basic_check(verbose=False)
```

```python
mysumstats.fix_id(fixid=True,forcefixid=True,overwrite=True,verbose=False)
```

```python
mysumstats.data
```

```python
sumstats2 = gl.Sumstats("../../0_sample_data/t2d_bbj.txt.gz", 
                             snpid="SNP",
                             chrom="CHR",
                             pos="POS",
                             ea="REF",
                             nea="ALT",
                             eaf="Frq",
                             beta="BETA",
                             se="SE",
                             p="P",
                             nrows=1000000,
                             verbose=False) 
```

```python
sumstats2.basic_check(verbose=False)
```

```python
pair = gl.SumstatsPair(mysumstats, sumstats2)
```

```python
pair.data
```

```python
pair.to_coloc(vcf = "/home/yunye/.gwaslab/EAS.ALL.split_norm_af.1kgp3v5.hg19.vcf.gz",n_cores=1, memory=10000)  
```

```python
pair.run_coloc_susie(r="/home/yunye/anaconda3/envs/r41/bin/Rscript",
                     types=("quant","cc"),
                     d1_args=",sdY=1", 
                     ns=(50000,150000))
```
