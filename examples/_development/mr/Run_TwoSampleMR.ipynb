## load gwaslab and sample data

```python
import sys
sys.path.insert(0,"/home/yunye/work/gwaslab/src")
import gwaslab as gl
```

```python
mysumstats = gl.Sumstats("../../0_sample_data/koges_bmi.tsv.gz", 
                             chrom="chrom",
                             pos="pos",
                             ea="alt",
                             nea="ref",
                             eaf="af",
                             beta="beta",
                             se="sebeta",
                             p="pval",nrows=1000000,
                            verbose=False) 
```

```python
mysumstats.basic_check(verbose=True)
```

```python
mysumstats.fix_id(fixid=True,forcefixid=True,overwrite=True,verbose=False)
```

```python
mysumstats.data
```

```python
mysumstats2 = gl.Sumstats("../../0_sample_data/t2d_bbj.txt.gz", 
                             snpid="SNP",
                             chrom="CHR",
                             pos="POS",
                             ea="REF",
                             nea="ALT",
                             eaf="Frq",
                             beta="BETA",
                             se="SE",
                             p="P",
                             nrows=1000000,
                             verbose=False) 
```

```python
mysumstats2.basic_check(verbose=False)
```

## create a SumstatsPair object

Allele and statistics in sumstats2 will be aligned using sumstats1 as reference

```python
pair = gl.SumstatsPair(mysumstats, mysumstats2)
```

```python
pair.data
```

### Run clumping

```python
pair.clump(vcf = gl.get_path("1kg_eas_hg19"))
```

```python
pair.clumps["clumps"]
```
