## Load sample data

```python
mysumstats = gl.Sumstats("t2d_bbj.txt.gz",
             snpid="SNP",
             chrom="CHR",
             pos="POS",
             ea="ALT",
             nea="REF",            
             beta="BETA",
             se="SE",
             p="P")
```

## Basic check

```python
mysumstats.basic_check()
```

## Get lead variants

```python
mysumstats.get_lead()
```

## Specifying different window sizes for extracting lead variants

```python
mysumstats.get_lead(windowsizekb=1000)
```

## Specifying different thresholds for extracting lead variants

```python
mysumstats.get_lead(sig_level=1e-20)
```

```python
mysumstats.plot_mqq(skip=2,cut=20,sig_level_lead=1e-20,anno=True)
```

## Use mlog10p instead of p

```python
#add MLOG10P column
mysumstats.fill_data(to_fill="MLOG10P")
```

```python
# gwaslab will use MLOG10P first by default, if MLOG10P is not avaiable, it will look for P column.
mysumstats.get_lead()
```
