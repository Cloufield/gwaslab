# Liftover

```python
import gwaslab as gl
```

```python
gl.show_version()
```

## Load sample data

```python
mysumstats = gl.Sumstats("../0_sample_data/t2d_bbj.txt.gz",
             snpid="SNP",
             chrom="CHR",
             pos="POS",
             ea="ALT",
             nea="REF",
             neaf="Frq",
             beta="BETA",
             se="SE",
             p="P",
             nrows=5000,
             verbose=False)
mysumstats.basic_check(verbose=False)
```

```python
mysumstats.data
```

## Liftover

```python
mysumstats.liftover(from_build="19", to_build="38")
```

```python
mysumstats.data
```
