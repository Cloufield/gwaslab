# Liftover

```python
import gwaslab as gl
```

## Load sample data

```python
mysumstats = gl.Sumstats("t2d_bbj.txt.gz",
             snpid="SNP",
             chrom="CHR",
             pos="POS",
             ea="ALT",
             nea="REF",
             neaf="Frq",
             beta="BETA",
             se="SE",
             p="P",
             nrows=500000,
             verbose=False)
mysumstats.basic_check(verbose=False)
```

```python
mysumstats.data
```

## Liftover

```python
mysumstats.liftover(n_cores=3, from_build="19", to_build="38")
```
