# Fix SNPID / Extract CHR, POS, NEA, EA from SNPID

```python
import gwaslab as gl
```

```python
gl.show_version()
```

## load sample data

```python
mysumstats = gl.Sumstats("../0_sample_data/toy_data/fixid.tsv",fmt="gwaslab")
```

```python
mysumstats.data
```

## Remove chr prefix and extract chr, pos, nea, ea from SNPID

```python
mysumstats.fix_id(fixchrpos=True, 
                  fixeanea =True, 
                  fixprefix=True)
```

```python
mysumstats.data
```

## run basic_check to clean the data

```python
mysumstats.basic_check(remove=True)
```

```python
mysumstats.data
```

## construct SNPID from CHR, POS, NEA, EA

```python
mysumstats.data = mysumstats.data.drop(columns=["SNPID"])
mysumstats.data
```

By default, gwaslab will check variant STATUS code and only fix the SNPID for those are aligned with reference genome. 
If not the data was not aligned, gwaslab will only use CHR and POS to create SNPID.

```python
mysumstats.fix_id(fixid=True)
mysumstats.data
```

You can force gwaslab to create SNPID using `forcefixid=True`. And`overwrite=True` will overwrite the existing SNPID

```python
mysumstats.fix_id(fixid=True, forcefixid=True, overwrite=True)
```

```python
mysumstats.data
```
